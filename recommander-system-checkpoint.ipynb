```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
# first step is to merge both of this table

movies = movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
movies.info()
```

```python
# geners
#keyword
# overview
# title
# cast
# crew
# id
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
# as there are only 3 rows of missing we ARE going to drop the rows
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
# preprocessing on genres coloumn
import ast
def convert(obj):
    li = []
    for i in ast.literal_eval(obj):
        li.append(i['name'])

    return li
```

```python
movies['genres'][0]
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
# featching actors
```

```python
def convertcast(obj):
    li = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            li.append(i['name'])
            counter +=1
        else:
            break

    return li
```

```python
movies['cast']=movies['cast'].apply(convert)
```

```python
def convertcast3(obj):
    li = []
    counter = 0
    for i in obj:
        if counter !=3:
            li.append(i)
            counter +=1
        else:
            break

    return li
```

```python
movies['cast']=movies['cast'].apply(convertcast3)
```

```python
movies.head(1)
```

```python
movies['crew'][0]
```

```python
# featiching director name

def featch_director(obj):
    li = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            li.append(i['name'])
            break

    return li
```

```python
movies['crew']=movies['crew'].apply(featch_director)
```

```python
# converting overview in the list format

movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:' '.join(x))
```

```python
new_df['tags'][1]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
st = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(st.stem(i))

    return " ".join(y)
```

```python
stem('creativity')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df.head(1)
```

```python
new_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
type(vectors)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
similarity[1]
```

```python
similarity[2]
```

```python
def recomande(movie):

    movie_index = new_df[new_df['title'] == movie].index[0]

    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(similarity[movie_index])),reverse=True,key=lambda x:x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recomande('Spectre')
```
